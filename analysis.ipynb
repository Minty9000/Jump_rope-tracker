```python
import sounddevice as sd
from scipy.io.wavfile import write

fs = 44100  # Sample rate (Hz)
seconds = 10  # Record for 10 seconds

print("ðŸŽ§ Recording... start jumping!")
recording = sd.rec(int(seconds * fs), samplerate=fs, channels=1, dtype='float64')
sd.wait()
print("âœ… Recording complete.")

write("jump_rope_hit.wav", fs, recording)
print("ðŸŽµ Saved as jump_rope_hit.wav")
```

```python
import matplotlib.pyplot as plt
from scipy.io import wavfile

fs, data = wavfile.read("jump_rope_hit.wav")

# If stereo, convert to mono
if len(data.shape) > 1:
    data = data.mean(axis=1)

time = range(len(data))
plt.figure(figsize=(12, 4))
plt.plot(time, data)
plt.title("Jump Rope Sound Waveform")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.show()
```

```python
import numpy as np

fft = np.fft.fft(data)
freqs = np.fft.fftfreq(len(fft), 1/fs)
magnitude = np.abs(fft)

plt.figure(figsize=(12, 4))
plt.plot(freqs[:len(freqs)//2], magnitude[:len(magnitude)//2])
plt.title("Frequency Spectrum of Rope Hit")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.show()
```
